```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('hate_crime_dataset.xlsx')
```

```python
df = df[['data_year', 'incident_date', 'bias_desc']]
df.head(5)
```

## 동기별 분류

```python
races = ['White', 'Black', 'Hispanic', 'Asian', 'Multiple Races']

df_Black = df.loc[df['bias_desc'].str.contains('Black')]
df_Hispanic = df.loc[df['bias_desc'].str.contains('Hispanic')]
df_Asian = df.loc[df['bias_desc'].str.contains('Asian')]
df_Multi = df.loc[df['bias_desc'].str.contains('Multiple Races')]
```

```python
df_black_tl = df_Black.groupby(['incident_date']).agg(cases = ('incident_date', 'size'))
df_black_tl = pd.DataFrame(df_black_tl).reset_index()

sns.set()
label = ['Cases']
plt.figure(figsize = (20, 8))
plt.plot(df_black_tl['incident_date'], df_black_tl['cases'], color='crimson')
plt.title("USA Black Hate Crimes", fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(labels = label, fontsize='large')
plt.show()
```

```python
df_latin_tl = df_Hispanic.groupby(['incident_date']).agg(cases = ('incident_date', 'size'))
df_latin_tl = pd.DataFrame(df_latin_tl).reset_index()

sns.set()
label = ['Cases']
plt.figure(figsize = (20, 8))
plt.plot(df_latin_tl['incident_date'], df_latin_tl['cases'], color='crimson')
plt.title("USA Asian Hate Crimes", fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(labels = label, fontsize='large')
plt.show()
```

```python
df_asian_tl = df_Asian.groupby(['incident_date']).agg(cases = ('incident_date', 'size'))
df_asian_tl = pd.DataFrame(df_asian_tl).reset_index()

sns.set()
label = ['Cases']
plt.figure(figsize = (20, 8))
plt.plot(df_asian_tl['incident_date'], df_asian_tl['cases'], color='crimson')
plt.title("USA Asian Hate Crimes", fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(labels = label, fontsize='large')
plt.show()
```
